```python
import  pandas as pd
```

```python
df = pd.read_csv("cleaned_train.csv")
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=False)  
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=False)
```

```python
category_columns = [
    "Ship Mode",
    "Segment",
    "Region",
    "Category",
    "Sub-Category"
]

for col in category_columns:
    df[col] = df[col].astype("category")
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
df = df.sort_values("Order Date")
```

```python
df.set_index("Order Date", inplace=True)
```

```python
daily_sales = df["Sales"].resample("D").sum()

plt.figure(figsize=(14, 4))
plt.plot(daily_sales, color="skyblue")
plt.title("Gnlk Sat Trendleri")
plt.xlabel("Tarih")
plt.ylabel("Toplam Sat")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
weekly_sales = df["Sales"].resample("W").sum()

plt.figure(figsize=(14, 4))
plt.plot(weekly_sales, color="orange")
plt.title("Haftalk Sat Trendleri")
plt.xlabel("Tarih")
plt.ylabel("Toplam Sat")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
monthly_sales = df["Sales"].resample("ME").sum()

plt.figure(figsize=(14, 4))
plt.plot(monthly_sales, color="green")
plt.title("Aylk Sat Trendleri")
plt.xlabel("Tarih")
plt.ylabel("Toplam Sat")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print(df.index)
```

```python
monthly_subcat_sales = df.groupby([pd.Grouper(freq="ME"), "Sub-Category"])["Sales"].sum().reset_index()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(14, 7))

sns.lineplot(
    data=monthly_subcat_sales,
    x="Order Date",
    y="Sales",
    hue="Sub-Category",
    palette="tab10"
)

plt.title("Alt Kategori Baznda Aylk Sat Trendleri")
plt.xlabel("Tarih")
plt.ylabel("Toplam Sat")
plt.legend(title="Sub-Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1,2,1)
sns.histplot(df["Sales"], bins=50, color="midnightblue")
plt.title("Sales Histogram Dalm")
plt.xlabel("Sales")
plt.ylabel("Frekans")
plt.grid(True)

plt.subplot(1, 2, 2)
sns.kdeplot(df["Sales"], color="darkorange", fill=True)
plt.title("Sales KDE Erisi")
plt.xlabel("Sales")
plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

```python
city_sales = df.groupby("City")["Sales"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=city_sales.values, y=city_sales.index, hue=city_sales.index, palette="viridis", legend=False)
plt.title("En ok Sat Yaplan lk 10 ehir")
plt.xlabel("Toplam Sat")
plt.ylabel("ehir")
plt.tight_layout()

plt.show()
```

```python
product_sales = df.groupby(["Sub-Category"], observed = False)["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=product_sales.values, y=product_sales.index, hue=product_sales.index, palette="coolwarm", legend=False)
plt.title("Alt Kategori Baznda Toplam Satlar")
plt.xlabel("Toplam Sat")
plt.ylabel("Alt Kategori")
plt.tight_layout()
plt.show()
```
